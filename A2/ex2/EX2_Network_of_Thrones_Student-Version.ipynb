# Task 2.1 - Creating a network
First, you will need to create a network from our dataset.
Our dataset consists of five csv files - one for each book - representing the co-occurence network of the characters in the books. Two characters are considered to co-occur if they are mentioned within 15 words of one another in a book.
(This kind of relationship may seem arbitrary, but co-occurence is a useful measure in Natural Language Processing.)

**a)** Import the first book and print some basic information about it. The column 'weight' corresponds to the number of co-occurences per pair. 

```python
import pandas as pd


# Read in the first book into a dataframe using the pandas library
book1 = pd.read_csv("dataset/book1.csv")

# Use two different functions to print information about the dataframe. 
# Common things you might want to know are: 
# * What columns and what datatypes do we have? 
# * How many rows are there? 
# * Do we have any missing values? 

# TODO
print(book1.head())

# TODO
print(book1.info())

# Describe the column 'weight'. Find out: What is its' mean? What is the minimum and maximum value?
# TODO
print(book1["weight"].describe())
print("Mean from column weight: " + str(book1["weight"].mean()))
print("Max from column weight: " + str(book1["weight"].max()))
print("Min from column weight: " + str(book1["weight"].min()))
       
# How many unique values does 'weight' take?
unique_values = book1["weight"].unique() 
print(f"{len(unique_values)} unique values in \'weight\''")

# How many unique characters (nodes) are there in our network? 
# Hint: you have to take both 'Source' and 'Target' into account.
unique_source = book1["Source"].unique()
unique_target = book1["Target"].unique()
unique_characters = len(unique_source) + len(unique_target)
print(f"{unique_characters} unique characters in the first book")
```

**b)** Create and fill graph. In order to do things with our data, we have to have it in graph form. 
1. Use the **networkx** library to create an empty graph for book1. Then populate it with the edges. Don't forget the edge weights! <br> *Hint:* It is generally advised to not iterate naively over dataframes as it is much slower than other methods. But, our dataset is quite small, so you can use iterrows().

(See the [networkx documentation](https://networkx.org/documentation/stable/reference/index.html) [4])

```python
import networkx as nx


# Create an empty graph 
G_book1 = nx.Graph()

# Add the edges of book1 to our graph
for _, row in book1.iterrows():
    G_book1.add_edge(row["Source"], row["Target"], weight=row["weight"])
```

2. We are going to need a graph for each book so that we can look at the evolution of our social network. Create and fill the graphs as before, and add them to the list books. 

```python
books = [G_book1]

# Create and fill a graph for each book. Add the graphs in the list books.  
book_fnames = ["dataset/book2.csv", "dataset/book3.csv", "dataset/book4.csv", "dataset/book5.csv"]
for book_fname in book_fnames:
    #TODO
    book = pd.read_csv(book_fname)
    G_book = nx.Graph()
    for _, row in book.iterrows():
        G_book.add_edge(row["Source"], row["Target"], weight=row["weight"])
    books.append(G_book)
```

### 2.2.1 Betweeness centrality ###
The more shortest baths go through this node, the higher its betweenness centrality. (Note that our dataset only contains undirected edges.)

#### a) The top 15 in the first and last books
Let's calculate the betweeness centrality of the first and last books. Then, for both books, sort for only the characters with the top 15 centralities. 

```python
# Calculate the betweeness centrality of book 1. You can use the pre-built function of networkx. Don't forget the weights!
bet_cen_book1 = nx.betweenness_centrality(G_book1, weight="weight")

# Calculate the betweeness centrality of book 5
bet_cen_book5 = nx.betweenness_centrality(books[4], weight="weight")

# Sort the dictionaries according to their degree centrality and store the top 15
sorted_bet_cen_book1 = sorted(bet_cen_book1.items(), key=lambda x:x[1], reverse=True)[:15]

# Sort the dictionaries according to their degree centrality and store the top 15
sorted_bet_cen_book5 = sorted(bet_cen_book5.items(), key=lambda x:x[1], reverse=True)[:15]

# Print out the top 15 of book1 and book5
print("Book 1:")
for character, centrality in sorted_bet_cen_book1:
    print(f"{character}: {centrality}")

print("Book 5:")
for character, centrality in sorted_bet_cen_book5:
    print(f"{character}: {centrality}")
```

#### b) The distribution of betweeness centralities

 ### 2.2.2 The evolution of character importance ###

#### **a)** The evolution of degree centrality #

We can see that the importance of Eddard Stark dies off as the book series progresses. With Jon Snow, there is a drop in the fourth book but a sudden rise in the fifth book.

In the following, we will look at the evolution of the top characters using different centrality measures, and see if we can find anything interesting.
 
Plot the evolution of degree centrality of this network over the five books. 
Take the evolution of the **top four characters** of every book and plot it.

What can you say about Stannis Baratheon? What might be the reason he gets the score he does? (You do not have to use knowledge from the books or series.) - Stannis`s scores vary a lot, he interacts (or is mentionend) a lot with other characters in Book 2 and 5 but not as much in the others, although he is a important character throughout all of them.

```python
# For the following evolution plots, use this colormap. 
# This will ensure that all top characters are marked with a unique color.
import matplotlib as mpl
cmap = mpl.colormaps['tab20']
```

```python
# Create a list of degree centrality of all the books. Hint: you can use the available networkx function. 
# Dont foret the weigths

# As networkx does not provide a function to compute weighted degree centrality, i will not use networkx
evol = []

def weighted_degree_centrality(book):
        weighted_degree_centrality = {}
        total_weight_in_book = sum(data["weight"] for _,_,data in book.edges(data=True))
        for node in book.nodes():
            # weighted degree centrality is the sum of the weights of the edges connected to a node.
            weighted_degree_centrality[node] = 0
            weighted_degree_centrality[node] = sum(data["weight"] for _,_,data in book.edges(node, data=True))
            # Divide by the max value of the total sum of weights for normalization (normal Degree Centrality divides by n - 1 as it is the max number of connections)
            weighted_degree_centrality[node] /= total_weight_in_book
        return weighted_degree_centrality


for book in books:
    evol.append(weighted_degree_centrality(book))
# Make a DataFrame from the list
degree_evol_df = pd.DataFrame(evol)
degree_evol_df = degree_evol_df.fillna(0) # Filling NaN values with 0 for better plotting

# Daenerys does not co-occur with anyone in the fourth book. Poor Dany! To be able to still plot her, we are going to set her centrality to 0.
degree_evol_df['Daenerys-Targaryen'][3] = 0

# Find the top 4 characters in every book. Hint: you can use sets to avoid duplicates. 
set_of_char = set()
for i in range(5):
    sorted_values = degree_evol_df.loc[i].sort_values(ascending=False)
    set_of_char.update(sorted_values.index[:4])
list_of_char = list(set_of_char)

# Print Stannis Baratheon
print(degree_evol_df["Stannis-Baratheon"])


# Plot the evolution of the top characters. You can use the provided colour map. Use a figure size of (13,7)
#TODO
colors = cmap(range(len(list_of_char)))

fig, ax = plt.subplots(figsize=(13, 7))
x_values = [1,2,3,4,5]
i = 0
for character in list_of_char:
    ax.plot(x_values, degree_evol_df[character],marker="o", label=character, color=colors[i])
    i += 1

ax.set_ylabel("Weighted Degree Centrality")
ax.set_xlabel("Books")

ax.set_xticks(range(1,6))

ax.legend()
```

#### **b)** Eigenvector-centrality ##

Eigenvector centrality takes into account the importance of a nodes' neighbours. You may have many friends, but if none of them are famous, chances are you are not either. If you want a more detailed explanation of eigenvector centrality (and a handy site for other centralities), refer [to this website](https://neo4j.com/docs/graph-data-science/current/algorithms/eigenvector-centrality/) [5].

How does Stannis score in this setting? - He scores overall lower although higher on the second book but still isnt enough to get top 4 (he actually gets 5 place), meaning that Stannis probably "had a lot of friends, but none of them (or few of them) were famous".

```python
# Create a list of eigenvector centrality of all the characters in all the books. Hint: The algorithm for eigenvector centrality is not guaranteed to converge, and stops after the default of 100 iterations. Select a higher number of max iterations to converge. 
evol = []
for book in books:
    eigenvector_centrality = nx.eigenvector_centrality(book, weight="weight", max_iter=500)
    evol.append(eigenvector_centrality)

# Make a DataFrame from the list
eigenvector_evol_df = pd.DataFrame(evol)
eigenvector_evol_df = eigenvector_evol_df.fillna(0) # Filling NaN values with 0 for better plotting

# Daenerys does not co-occur with anyone in the fourth book. Poor Dany! To be able to still plot her, we are going to set her centrality to 0.
eigenvector_evol_df['Daenerys-Targaryen'][3] = 0

# Print Stannis Baratheon
print(eigenvector_evol_df["Stannis-Baratheon"])

# Find the top 4 characters in every book
set_of_char = set()
for i in range(5):
    sorted_values = eigenvector_evol_df.loc[i].sort_values(ascending=False)
    top_chars = sorted_values.index[:4]  # Extract the top 4 column names
    set_of_char.update(top_chars)  # Add the column names to the set
list_of_char = list(set_of_char)  # Convert the set to a list

# Plot the evolution of the top characters. You can use the provided colour map.  Use a figure size of (13,7)
#TODO
colors = cmap(range(len(list_of_char)))

fig, ax = plt.subplots(figsize=(13, 7))
x_values = [1,2,3,4,5]
i = 0
for character in list_of_char:
    ax.plot(x_values, eigenvector_evol_df[character],marker="o", label=character, color=colors[i])
    i += 1

ax.set_ylabel("Eigenvector Centrality")
ax.set_xlabel("Books")

ax.set_xticks(range(1,6))

ax.legend()
```

#### **c)** PageRank Centrality ###

PageRank was the initial way Google ranked web pages. It evaluates the inlinks and outlinks of webpages in the world wide web, which is, essentially, a directed network. Let's look at the importance of characters in the Game of Thrones network according to PageRank.

What does PageRank measure in our co-occurence network? - It measures how important the characters are based on many interactions (and how meaningfull they were) with important characters one has, in more detail: It evaluates direct and indirect connections, incorporates edge weights and takes into consideration if the character is connected to people that have high PageRanks.

```python
# Create a list of pagerank of all the characters in all the books
evol = []
for book in books:
    pagerank_centrality = nx.pagerank(book, weight="weight")
    evol.append(pagerank_centrality)

# Make a DataFrame from the list
pagerank_evol_df = pd.DataFrame(evol)
pagerank_evol_df = pagerank_evol_df.fillna(0)  # Filling NaN values with 0 for better plotting

# Daenerys does not co-occur with anyone in the fourth book. Poor Dany! To be able to still plot her, we are going to set her centrality to 0.
pagerank_evol_df['Daenerys-Targaryen'][3] = 0

# Find the top 4 characters in every book
set_of_char = set()
for i in range(5):
    sorted_values = pagerank_evol_df.loc[i].sort_values(ascending=False)
    top_chars = sorted_values.index[:4]  # Extract the top 4 column names
    set_of_char.update(top_chars)  # Add the column names to the set
list_of_char = list(set_of_char)  # Convert the set to a list

# Plot the evolution of the top characters. You can use the provided colour map. Use a figure size of (13,7)
#TODO
colors = cmap(range(len(list_of_char)))

fig, ax = plt.subplots(figsize=(13, 7))
x_values = [1,2,3,4,5]
i = 0
for character in list_of_char:
    ax.plot(x_values, pagerank_evol_df[character],marker="o", label=character, color=colors[i])
    i += 1

ax.set_ylabel("PageRank Centrality")
ax.set_xlabel("Books")

ax.set_xticks(range(1,6))

ax.legend()
```

#### Correlation between different measures ###

Stannis, Jon Snow, and Daenerys are the most important characters in the fifth book according to PageRank. Eddard Stark follows a similar curve but for degree centrality and betweenness centrality: He is important in the first book but dies into oblivion over the book series.

We have seen three different measures to calculate the importance of a node in a network, and all of them tells us something about the characters and their importance in the co-occurrence network. We see some names pop up in all four measures. Can we explain this? Let's calculate the correlation between measures. 

#### **d)** Calculating correlations
Let's look at the correlation between PageRank, eigenvector centrality, betweenness centrality and degree centrality for the fifth book using Pearson correlation.[3]

```python
# Create a list of pagerank, betweenness centrality, degree centrality and eigenvector of all the characters in the fifth book.
measures = [nx.pagerank(books[4], weight='weight'), nx.betweenness_centrality(books[4], weight='weight'), weighted_degree_centrality(books[4]),nx.eigenvector_centrality(books[4], weight='weight', max_iter=500)]

# Name the columns and rows accordingly
names = ['Pagerank','Betweenness','Degree','Eigenvector']

# Create a dataframe from measures
cor = pd.DataFrame(measures, index=names).transpose()
print(cor.head())

# Calculate the correlation. Hint: you can use corr().
#TODO
corr = cor.corr()
print(corr)
```
