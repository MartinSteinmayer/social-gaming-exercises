## Task 1.1: Python Rhombus
**a)** **Create** a function which takes a number of levels (N) and **prints** a rhombus that looks like this for N = 4:

```python
 # 
###
###
 #
```

for N = 5:

```python
  #
 ###
#####
 ###
  #
```

**Hints:**
- Do not forget the spaces left of the rhombus, except on the widest floor. In the example above 2 spaces to the left and at the peak.  
- In order to execute a code cell, press Shift + Enter.

```python
def printRhombus(N):
    #TODO: Create a Rhombus
    i = 1
    spaces = (N-1)//2
    while i < N:
        print(spaces*" " + i*"#")
        i += 2
        spaces -= 1

    spaces = 0
    if (N % 2 != 0):
        print(N*"#")
        spaces = 1
    i -= 2
    while i > 0:
        print(spaces*" " + i*"#")
        i -= 2
        spaces += 1
        
        
printRhombus(10)
```

## Task 1.2 : The Marvel Universe is introducing Social Computing
In social computing research, we need powerful tools to create, manipulate and display graphs. Luckily, there is a plethora of tools and libaries for that. 
NetworkX is a Python package for the creation, manipulation, and study of the structure, dynamics, and functions of complex networks.

Especially for the second exercise sheet, we are going to use [**NetworkX**](https://networkx.github.io) [1]. It provides rich graph data structures and many out-of-the-box functions to process graphs and calculate different metrics. The tasks below should make you familiar with the library.<br>
Please consult the [reference](https://networkx.github.io/documentation/stable/reference/index.html) [2] and the [tutorial](https://networkx.github.io/documentation/stable/tutorial.html) [3].

To give you a short background on graph visualization, it is the research area in mathematics/computer science concerned with drawing graphs. It has applications in many fields, one of them is social computing. The quality of graph visualization is measured based on certain criteria, for example crossing minimization and bend minimization. There are many graph drawing algorithms that vary in their quality according to the graph's application and size. One technique is to draw graphs by using physical analogies.<br>
The basic idea of this technique is to associate edges between graph nodes with physical forces acting upon the nodes and computing an energy minimum. By setting off the dynamics that is induced by the forces, the graph will finally settle into a natural optimal display. A famous algorithm that implements this technique is the **Fruchtermann-Rheingold** algorithm. Its basic idea is to replace the graph edges with mechanical springs, and let the springs move the system to a minimal energy state.

**Import** the necessary libraries for this exercise.

```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt
```

**Read** the csv files into pandas dataframes.

```python
# Read the csv into pandas DataFrames
df_edges = pd.read_csv("MarvelAppearences.csv")
df_edges
```

**a)** To get to see which characters appear together merge the Dataframe together with itself. This is not unlike the join operation in SQL. 

```python
# TODO: Merge charactes
# Hint: Use a left join
df_merged = df_edges.merge(df_edges, on="Appearence")
df_merged
```

**b)** Next we will group together the same characters on both sides to get the amount of times these characters appeared together. This will be used as the weights later. 

```python
df_merged2 = df_merged.groupby(["Character_x", "Character_y"]).count().sort_values(by="Appearence", ascending=False) # Sorting at the end for better visualization
df_merged2 = df_merged2.rename(columns={"Appearence" : "Weights"}) # Renaming for better understanding
# TODO: Group the Characters together

df_merged2 = df_merged2.reset_index()  #To transfrom back the columns to Character_x and Character_y
df_merged2
```

**c)** Now we are only interested in characters who have appeared at least 100 times together. Select those. Additionally delete rows of characters with references to themselves. 

```python
# TODO: Select the rows of characters with less than 100 appearances
indexNames = df_merged2[df_merged2['Weights'] < 100].index

# Delete these row indices from the dataframe
df_merged2.drop(indexNames, inplace=True)

# TODO: Select self-referential characters

indexNames2 = df_merged2[df_merged2["Character_x"] == df_merged2["Character_y"]].index

# Delete these rows
df_merged2.drop(indexNames2, inplace=True)
df_merged2
```

**d)** Now you have to **include your alter ego into the network**. Create a pandas Series with your name, and weights. Connect yourself to THANOS with weight 345.

```python
# TODO:
# Create a series for your character who is connected to THANOS 345 times and add it to the dataframe
martin = pd.Series(["Martin Steinmayer", "THANOS", 345], index=["Character_x", "Character_y", "Weights"])

# TODO: 
# Append the newly created series to the pandas data frame
df_merged3 = df_merged2
df_merged3.loc[len(df_merged2)] = martin

# Create the graph from the dataframe
graph = nx.from_pandas_edgelist(df_merged3, source="Character_x", target="Character_y", edge_attr=True)
df_merged3
```

**e)** **Draw** the resulting graph with the given options. Choose 2 [layout](https://networkx.github.io/documentation/stable/reference/drawing.html) [6] options that seem the most suitable for the data. Briefly discuss why you chose these over the others.

```python
# Set the edge color according to the weight
edges,weights = zip(*nx.get_edge_attributes(graph,'Weights').items())

# Style the graph
options = {
    "font_size" : 14,
    "font_color" : '#552222',
    "node_color" : '#22FF22',
    "width" : 5.0,
    "edgelist" : edges,
    "edge_color" : weights,
    "edge_cmap" : plt.cm.Blues
}

plt.figure(1,figsize=(40,40)) 

# TODO: plot the graph
kamada_kawai = nx.kamada_kawai_layout(graph, weight=weights)
nx.draw(graph, pos=kamada_kawai, **options)
plt.title("Kamada-Kawai Layout", {"fontsize": 50})
plt.show()


plt.figure(2,figsize=(40,40)) 
spring_layout = nx.spring_layout(graph, k=0.3)
nx.draw(graph, pos=spring_layout, **options)
plt.title("Fruchterman-Reingold Layout", {"fontsize": 50})
plt.show()
```
